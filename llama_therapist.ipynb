```python
## Seeding RNGs for reproducible training
import random
import numpy as np
import torch
SEED = 42

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False
```

```python
from unsloth import FastLanguageModel
import torch
max_seq_length = 150 # Choose any! We auto support RoPE Scaling internally!
dtype = None # None for auto detection. Float16 for Tesla T4, V100, Bfloat16 for Ampere+
load_in_4bit = True # Use 4bit quantization to reduce memory usage. Can be False.


model, tokenizer = FastLanguageModel.from_pretrained(
    model_name = "unsloth/Llama-3.2-3B-Instruct", # or choose "unsloth/Llama-3.2-1B-Instruct"
    max_seq_length = max_seq_length,
    dtype = dtype,
    load_in_4bit = load_in_4bit,
)
```

```python
model = FastLanguageModel.get_peft_model(
    model,
    r = 16, # Choose any number > 0 ! Suggested 8, 16, 32, 64, 128
    target_modules = ["q_proj", "k_proj", "v_proj", "o_proj",
                      "gate_proj", "up_proj", "down_proj",],
    lora_alpha = 16,
    lora_dropout = 0, # Supports any, but = 0 is optimized
    bias = "none",    # Supports any, but = "none" is optimized
    # [NEW] "unsloth" uses 30% less VRAM, fits 2x larger batch sizes!
    use_gradient_checkpointing = "unsloth", # True or "unsloth" for very long context
    random_state = 3407,
    use_rslora = False,  # We support rank stabilized LoRA
    loftq_config = None, # And LoftQ
)
```

```python
# Adjust to ensure compatibility without relying on `xformers.attn_bias`
if not hasattr(model, "attn_bias"):
    model.attn_bias = None  # Explicitly set to None as a placeholder

# Save the original forward method
original_forward = model.__class__.forward

# Define a patched forward method
def patched_forward(self, *args, **kwargs):
    if "attn_bias" in kwargs and kwargs["attn_bias"] is None:
        kwargs.pop("attn_bias")  # Remove the problematic key if it exists
    return original_forward(self, *args, **kwargs)

# Assign the patched forward method
model.__class__.forward = patched_forward

# Update the model configuration for compatibility
if hasattr(model.config, "attention_bias"):
    model.config.attention_bias = False
```

```python
# Function to preprocess each row
def preprocess_conversations(row):
    """
    Extracts human input and GPT responses from the conversations column of a single row.
    Returns a list of dictionaries with 'input' and 'response' pairs.
    """
    pairs = []
    conversations = row["conversations"]  # Already a list of dictionaries
    for i in range(0, len(conversations) - 1, 2):  # Iterate in steps of 2 (human -> gpt)
        if conversations[i]["from"] == "human" and conversations[i + 1]["from"] == "gpt":
            pairs.append({
                "input": conversations[i]["value"],
                "response": conversations[i + 1]["value"]
            })
    return pairs

# Preprocess the entire dataset
all_pairs = []
for _, row in df.iterrows():  # Iterate over all rows in the DataFrame
    all_pairs.extend(preprocess_conversations(row))  # Collect input-response pairs from each row

# Create a new DataFrame with input-response pairs
processed_df = pd.DataFrame(all_pairs)

# Save or use the processed DataFrame
processed_df.to_csv("processed_dataset.csv", index=False)  # Save for future use
processed_df.head()  # Show the first few rows of the processed data
```

```python
from unsloth.chat_templates import get_chat_template
from datasets import Dataset
import pandas as pd

# Load the processed dataset
processed_df = pd.read_csv('processed_dataset.csv') # Ensure this is the output from the preprocessing function

# Convert the processed dataset back to the required format
# Combine 'input' and 'response' into a 'conversations' column with a list of dictionaries
processed_df["conversations"] = processed_df.apply(
    lambda row: [
        {"from": "human", "value": row["input"]},
        {"from": "gpt", "value": row["response"]}
    ],
    axis=1
)

# Drop unnecessary columns to retain only 'id' and 'conversations'
processed_df = processed_df[["conversations"]]

from sklearn.model_selection import train_test_split
from datasets import Dataset

# Split dataset into train and validation sets
train_df, val_df = train_test_split(processed_df, test_size=0.005, random_state=42)

# Convert to HuggingFace datasets
train_dataset = Dataset.from_pandas(train_df)
val_dataset = Dataset.from_pandas(val_df)

print("Training dataset size:", len(train_dataset))
print("Validation dataset size:", len(val_dataset))

# Set up tokenizer chat template
tokenizer = get_chat_template(
    tokenizer,
    chat_template="llama-3.1",  # Update if your model uses a different template
)

# Formatting function for prompts
def formatting_prompts_func(examples):
    convos = examples["conversations"]
    texts = [
        tokenizer.apply_chat_template(convo, tokenize=False, add_generation_prompt=False)
        for convo in convos
    ]
    return {"text": texts}
```

```python
from unsloth.chat_templates import standardize_sharegpt
train_dataset = standardize_sharegpt(train_dataset)
train_dataset = train_dataset.map(formatting_prompts_func, batched = True,)
val_dataset = standardize_sharegpt(val_dataset)
val_dataset = val_dataset.map(formatting_prompts_func, batched = True,)
```

```python
from trl import SFTTrainer
from transformers import TrainingArguments, DataCollatorForSeq2Seq
from unsloth import is_bfloat16_supported

trainer = SFTTrainer(
    model = model,
    tokenizer = tokenizer,
    train_dataset = train_dataset,
    dataset_text_field = "text",
    max_seq_length = max_seq_length,
    data_collator = DataCollatorForSeq2Seq(tokenizer = tokenizer),
    dataset_num_proc = 2,
    packing = False, # Can make training 5x faster for short sequences.
    args = TrainingArguments(
        per_device_train_batch_size = 2,
        gradient_accumulation_steps = 4,
        warmup_steps = 5,
        num_train_epochs = 1, # Set this for 1 full training run.
        max_steps = 48,
        learning_rate = 2e-4,
        fp16 = not is_bfloat16_supported(),
        bf16 = is_bfloat16_supported(),
        logging_steps = 1,
        optim = "adamw_8bit",
        weight_decay = 0.01,
        lr_scheduler_type = "linear",
        seed = 3407,
        output_dir = "outputs",
        report_to = "none", # Use this for WandB etc
    ),

)
```

```python
from unsloth.chat_templates import train_on_responses_only
trainer = train_on_responses_only(
    trainer,
    instruction_part = "<|start_header_id|>user<|end_header_id|>\n\n",
    response_part = "<|start_header_id|>assistant<|end_header_id|>\n\n",
)
```

```python
from torch.utils.data import DataLoader

# Define a custom collate function for batching
def collate_fn(batch):
    input_texts = []
    reference_texts = []

    for example in batch:
        conversation = example["conversations"]
        input_texts.append(conversation[0]["content"])  # User's input
        reference_texts.append(conversation[1]["content"])  # GPT's expected response

    return input_texts, reference_texts

# Create a DataLoader
val_dataloader = DataLoader(
    val_dataset,
    batch_size=8,  # Adjust based on GPU memory and model size
    shuffle=False,  # No need to shuffle for validation
    collate_fn=collate_fn
)
```

```python
from tqdm import tqdm
from transformers import TextStreamer
import evaluate

# Load BLEU and ROUGE metrics
rouge = evaluate.load("rouge")
meteor = evaluate.load("meteor")
bertscore = evaluate.load("bertscore")

def evaluate_model_with_dataloader(model, tokenizer, val_dataloader,  rouge):
    """
    Evaluate the model using a DataLoader for the validation dataset.
    """
    predictions = []
    references = []

    # Iterate over the DataLoader
    for batch in tqdm(val_dataloader, desc="Evaluating"):
        input_texts, reference_texts = batch

        # Prepare inputs for the model
        messages = [{"role": "user", "content": text} for text in input_texts]
        inputs = tokenizer.apply_chat_template(
            messages,
            tokenize=True,
            add_generation_prompt=True,
            return_tensors="pt",
            padding=True,
            truncation=True,
        )

        # Ensure correct device placement
        if isinstance(inputs, torch.Tensor):
            input_ids = inputs.to("cuda")
            attention_mask = torch.ones_like(input_ids, dtype=torch.long).to("cuda")
        else:
            input_ids = inputs["input_ids"].to("cuda")
            attention_mask = inputs["attention_mask"].to("cuda")

        # Generate outputs in batch
        outputs = model.generate(
            input_ids=input_ids,
            attention_mask=attention_mask,
            max_new_tokens=150,
            use_cache=True,
            temperature=1.5,
            min_p=0.1,
        )

        # Decode predictions
        decoded_predictions = [tokenizer.decode(output, skip_special_tokens=True) for output in outputs]

        # Append predictions and references
        predictions.extend(decoded_predictions)
        references.extend([[ref] for ref in reference_texts[:len(decoded_predictions)]])  # Align with predictions


    rouge_scores = rouge.compute(predictions=predictions, references=references)
    meteor_score = meteor.compute(predictions=predictions, references=references)
    bert_score = bertscore.compute(predictions=predictions, references=references, lang="en")


    print(f"BERTScore: {bert_score['f1']}")
    print(f"METEOR Score: {meteor_score['meteor']}")
    print(f"ROUGE-1: {rouge_scores['rouge1']:.4f}")
    print(f"ROUGE-2: {rouge_scores['rouge2']:.4f}")
    print(f"ROUGE-L: {rouge_scores['rougeL']:.4f}")

    return {

        "rouge1": rouge_scores["rouge1"],
        "rouge2": rouge_scores["rouge2"],
        "rougeL": rouge_scores["rougeL"],
    }

# Example usage
evaluation_results = evaluate_model_with_dataloader(model, tokenizer, val_dataloader, rouge)


```
